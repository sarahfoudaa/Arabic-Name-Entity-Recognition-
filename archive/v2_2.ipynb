```python
def entities(test):
  no_of_sen =0
  df_sen = []
  s = ''
  entities_per_line = []
  entities = ['O', 'B-PERS', 'B-LOC', 'I-PERS', 'B-ORG', 'I-ORG', 'B-MISC', 'I-MISC', 'I-LOC']
  ss = ''
  for i in range (len(test)):
    if test[0][i]!='':
      s = s +" "+ test[0][i]
      no_of_sen = no_of_sen+1
      if test[1][i] in entities and test[1][i] not in ss:
        ss = ss +" "+ test[1][i]
    else:
      df_sen.append(s)
      s=''
      entities_per_line.append(ss)
      ss=''
  df_sen = pd.DataFrame(df_sen, columns=['sentence'])
  entities_per_line = pd.DataFrame(entities_per_line, columns=['entities'])
  sen_entities = pd.DataFrame().assign(sentence=df_sen['sentence'], entities=entities_per_line['entities'])
  return sen_entities
```

```python
def remove_punc(test):
  punc = '،-!"#$%&\'()*+,-/:;<=>?@[\\]^_`{|}~'
  punctuation = set(punc)
  # Create boolean mask to identify rows containing punctuation
  mask = test[0].apply(lambda x: any(char in punctuation for char in x))

  # Filter original DataFrame using boolean mask
  new_df = test[mask]
  test_wo_punc = test[mask==False]
  test_wo_punc = test_wo_punc.reset_index(drop=True)
  return test_wo_punc
```

```python
#receiving dataframe w/o punc
def analyse(test):
  sen_id = []
  sen_no = 0
  lendf = len(test)
  sum_of_lenths = 0
  summ = []
  len_of_sen = []
  c=0
  for i in range(lendf):
    if test[0][i]!='':
      #print(i)
      sen_id.append(sen_no)
      #counter to count words in a sentence
      c = c+1
    else:
      sen_no = sen_no + 1
      len_of_sen.append(c)
      c=0

  sum_of_lenths  = sum(len_of_sen)
  no_of_sen = len(len_of_sen)
  avg = sum_of_lenths / no_of_sen
  minn = min(i for i in len_of_sen if i > 0)
  maxx = max(len_of_sen)

  for i in range (len(len_of_sen)):
    summ.append((len_of_sen[i] - avg)**2)

  std =  math.sqrt((sum(summ))/len(len_of_sen))
  sen_id = pd.DataFrame(sen_id, columns=['sentence_id'])
  return minn, maxx, avg, std, sen_id
```
